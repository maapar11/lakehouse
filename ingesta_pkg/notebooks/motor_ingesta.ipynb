# Motor de Ingesta

En este cuaderno configuraremos el motor de la Ingesta./
Realizaremos las llamadas a los notebooks que contienen las ingestas de datos.

Creamos dos variables con las rutas de los notebooks:

```python
notebook_batch_path = "./ingesta_datos_batch"
notebook_streaming_path = "./ingesta_datos_streaming"
```

Generamos la configuración Lakehouse

```python
account = spark.conf.get("adls.account.name")

landing_container = f"abfss://landing-tarea@{account}.dfs.core.windows.net"
lakehouse_container = f"abfss://lakehouse-tarea@{account}.dfs.core.windows.net"

landing_path = landing_container
raw_path = f"{lakehouse_container}/raw"
bronze_path = f"{lakehouse_container}/bronze"
```

Ejecutamos notebook de ingesta batch y streaming.

```python
try:
    dbutils.notebook.run(notebook_batch_path, timeout_seconds=0)
    print("✅ Ingesta batch completada")
except Exception as e:
    print("⚠️ Error en la ingesta batch:", e)
```

```python
try:
    dbutils.notebook.run(notebook_streaming_path, timeout_seconds=0)
    print("✅ Ingesta streaming lanzada")
except Exception as e:
    print("⚠️ Error en la ingesta streaming:", e)
```
